```python
import pandas as pd
```

```python
df = pd.read_csv("superstore.csv")
df.head()
```

```python
df.info()
```

```python
# Summary statistics using .describe()
df.describe(include='all')
```

```python
#check for duplicates
duplicate_count = df.duplicated().sum()
print("duplicates : ", duplicate_count)
```

```python
#check for null (missing) values
df.isnull().sum()
```

```python
#trim string columns for unwanted spaces
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# Renaming columns according to snake casing for better readability and documentation
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('-','_')
df.head()
```

```python
# Drop Row ID (Redundant for SQL)
df = df.drop(columns=['row_id'])
df
```

```python
from sqlalchemy import create_engine, text
import psycopg2
```

```python
# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details
username = "postgres"      # default user
password = "root"          # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "Sales"         # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "orders"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```

```python
# Select all data
data = pd.read_sql(f"SELECT * FROM {table_name}", engine)
data
```
